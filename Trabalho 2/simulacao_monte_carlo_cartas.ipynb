# Simulação Monte Carlo

## Uma caixa contém 48 cartas, distribuídas igualmente em 4 tipos:

### 12 cartas do tipo A  
### 12 cartas do tipo B  
### 12 cartas do tipo C  
### 12 cartas do tipo D  

### São retiradas 7 cartas aleatoriamente sem reposição.

(a) Qual é a probabilidade de obter pelo menos uma carta do tipo B e uma do tipo D?  
(b) Refaça o cálculo supondo que as cartas sejam retiradas com reposição  
(c) Explique qualitativamente a diferença entre os dois resultados

```python
import numpy as np
import math

# ============================================================
# QUESTÃO 5 — Cartas (sem reposição e com reposição)
# ============================================================

def comb(n, k):
    """Combinação nCk usando math.comb (Python 3.8+)"""
    return math.comb(n, k)

def prob_exata_sem_reposicao():
    """
    (a) Sem reposição:
    P(pelo menos uma B e uma D) = 1 - P(sem B) - P(sem D) + P(sem B e sem D)

    Total: C(48,7)
    Sem B: escolhe 7 dentre as 36 (A,C,D) => C(36,7)
    Sem D: idem => C(36,7)
    Sem B e sem D: escolhe 7 dentre as 24 (A,C) => C(24,7)
    """
    total = comb(48, 7)
    sem_B = comb(36, 7)
    sem_D = comb(36, 7)
    sem_B_sem_D = comb(24, 7)

    p = 1 - (sem_B/total) - (sem_D/total) + (sem_B_sem_D/total)
    return p

def prob_exata_com_reposicao():
    """
    (b) Com reposição:
    Cada retirada é independente e:
      P(B)=12/48=1/4
      P(D)=1/4

    P(pelo menos uma B e uma D) = 1 - P(sem B) - P(sem D) + P(sem B e sem D)
    Sem B: (3/4)^7
    Sem D: (3/4)^7
    Sem B e sem D: (1 - P(B) - P(D))^7 = (1 - 1/4 - 1/4)^7 = (1/2)^7
    """
    p = 1 - 2*(3/4)**7 + (1/2)**7
    return p

def mc_sem_reposicao(R=200000, seed=20260406):
    """
    Monte Carlo sem reposição:
    Simula retirar 7 cartas de um baralho com 48 (12 de cada tipo).
    Conta frequência de eventos: "tem pelo menos uma B e uma D".
    """
    rng = np.random.default_rng(seed)

    # Representação: 0=A, 1=B, 2=C, 3=D
    baralho = np.array([0]*12 + [1]*12 + [2]*12 + [3]*12)

    sucesso = 0
    for _ in range(R):
        amostra = rng.choice(baralho, size=7, replace=False)
        tem_B = np.any(amostra == 1)
        tem_D = np.any(amostra == 3)
        if tem_B and tem_D:
            sucesso += 1

    return sucesso / R

def mc_com_reposicao(R=200000, seed=20260406):
    """
    Monte Carlo com reposição:
    Cada retirada é i.i.d. com P(A)=P(B)=P(C)=P(D)=1/4.
    """
    rng = np.random.default_rng(seed)

    sucesso = 0
    for _ in range(R):
        amostra = rng.integers(0, 4, size=7)  # 0..3 equiprováveis
        tem_B = np.any(amostra == 1)
        tem_D = np.any(amostra == 3)
        if tem_B and tem_D:
            sucesso += 1

    return sucesso / R

def questao5():
    """
    Executa cálculos exatos e simulações MC.
    """
    print("\n=== QUESTÃO 5 ===")

    p_ex_sem = prob_exata_sem_reposicao()
    p_ex_com = prob_exata_com_reposicao()

    print(f"(Exato) Sem reposição: {p_ex_sem:.6f}")
    print(f"(Exato) Com reposição: {p_ex_com:.6f}")

    # Simulações (ajuste R se quiser mais precisão)
    R = 200000
    p_mc_sem = mc_sem_reposicao(R=R)
    p_mc_com = mc_com_reposicao(R=R)

    print(f"(MC) Sem reposição (R={R}): {p_mc_sem:.6f}")
    print(f"(MC) Com reposição (R={R}): {p_mc_com:.6f}")

    # (c) Diferença qualitativa (comentário):
    # Sem reposição: as cartas retiradas mudam a composição restante,
    # criando dependência entre as retiradas (correlação).
    # Com reposição: retiradas independentes; probabilidades fixas em cada passo.

# Executa
if __name__ == "__main__":
    questao5()
```
