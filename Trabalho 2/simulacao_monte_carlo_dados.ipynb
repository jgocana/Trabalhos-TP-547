(a) Utilize simulação de Monte Carlo para estimar a probabilidade de que a soma dos

(b) Compare o valor obtido com o valor teórico  
(c) Discuta como o número de simulações afeta a precisão do resultado

```python
import numpy as np
import math

# ============================================================
# QUESTÃO 6 — Dois dados: P(soma > 9)
# ============================================================

def prob_teorica_soma_maior_9():
    """Valor teórico: 6/36 = 1/6"""
    return 1/6

def mc_dois_dados(N, rng):
    """
    (a) Simula N lançamentos de dois dados justos.
    Retorna estimativa de P(soma > 9).
    """
    # Dados com valores 1..6
    d1 = rng.integers(1, 7, size=N)
    d2 = rng.integers(1, 7, size=N)

    soma = d1 + d2
    evento = (soma > 9)  # booleano
    return evento.mean()

def questao6(seed=20260406):
    """
    (a) Estima via MC
    (b) Compara com teórico
    (c) Mostra como N afeta o erro (opcional: vários N)
    """
    rng = np.random.default_rng(seed)
    p_teo = prob_teorica_soma_maior_9()

    print("\n=== QUESTÃO 6 ===")
    print(f"Probabilidade teórica: {p_teo:.6f}")

    # Testa vários N para ver convergência
    Ns = [10**2, 10**3, 10**4, 10**5, 10**6]

    for N in Ns:
        p_hat = mc_dois_dados(N, rng)
        erro = abs(p_hat - p_teo)
        print(f"N={N:>8d} | p_hat={p_hat:.6f} | erro_abs={erro:.6e}")

    # (c) Comentário:
    # Em Monte Carlo, o erro típico decresce aproximadamente como 1/sqrt(N).
    # Ou seja, para reduzir o erro em 10x, normalmente precisa ~100x mais simulações.

# Executa
if __name__ == "__main__":
    questao6()
```
