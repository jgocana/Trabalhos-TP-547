```python
import numpy as np
import time

# ============================================================
# QUESTÃO 4 — Aceitação-Rejeição para f(x)=3x^2 em [0,1]
# ============================================================

def amostrar_ar_f_3x2(n, rng):
    """
    (a) Implementa aceitação-rejeição para f(x)=3x^2 em [0,1].

    Proposta: g(x)=Uniforme(0,1)
    g(x)=1 em [0,1]
    M = max f(x)/g(x) = 3

    Algoritmo:
      repetir:
        Y ~ U(0,1)
        U ~ U(0,1)
        aceita se U <= f(Y)/(M*g(Y)) = 3Y^2/3 = Y^2
    Retorna:
      amostras aceitas e número total de propostas (para taxa de aceitação)
    """
    M = 3.0

    amostras = []
    propostas = 0

    while len(amostras) < n:
        Y = rng.random()   # proposta ~ Uniforme(0,1)
        U = rng.random()   # uniforme para aceitar/rejeitar
        propostas += 1

        # probabilidade de aceitação = Y^2
        if U <= Y**2:
            amostras.append(Y)

    amostras = np.array(amostras)
    taxa_aceitacao = n / propostas
    return amostras, taxa_aceitacao, propostas

def amostrar_inversa_f_3x2(n, rng):
    """
    (c) Método da inversa:
      f(x)=3x^2, 0<=x<=1
      F(x)=∫_0^x 3t^2 dt = x^3
      Se U~U(0,1), então X=U^(1/3) tem essa distribuição
    """
    U = rng.random(n)
    X = U ** (1.0/3.0)
    return X

def questao4_comparar(n=100000, seed=20260406):
    """
    (b) Calcula taxa de aceitação (empírica e teórica).
    (c) Compara AR com inversa (tempo e eficiência).
    """
    rng = np.random.default_rng(seed)

    print("\n=== QUESTÃO 4 ===")
    print("Gerar amostras de f(x)=3x^2 em [0,1]")

    # --- Aceitação-Rejeição
    t0 = time.perf_counter()
    x_ar, taxa, propostas = amostrar_ar_f_3x2(n, rng)
    t1 = time.perf_counter()

    # Taxa teórica = 1/M = 1/3
    taxa_teorica = 1/3

    print(f"(AR) n={n} | propostas={propostas} | taxa_empírica={taxa:.4f} | taxa_teórica={taxa_teorica:.4f}")
    print(f"(AR) tempo: {(t1-t0):.4f} s")

    # --- Inversa
    rng2 = np.random.default_rng(seed)  # mesma semente para comparação justa
    t2 = time.perf_counter()
    x_inv = amostrar_inversa_f_3x2(n, rng2)
    t3 = time.perf_counter()

    print(f"(Inversa) n={n} | tempo: {(t3-t2):.4f} s")

    # (d) Eficiência: inversa costuma ser muito mais rápida (não rejeita nada)
    # AR desperdiça propostas (~2/3 rejeitadas, pois taxa ~ 1/3).

# Executa
if __name__ == "__main__":
    questao4_comparar()
```
