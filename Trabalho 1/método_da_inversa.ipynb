```python
"""
Método da Inversa

Dada a função densidade de probabilidade:
f(x) = 2x, 0 ≤ x ≤ 1

O objetivo é:
(a) Verificar se é uma pdf válida
(b) Determinar a CDF
(c) Gerar amostras via transformação inversa
(d) Comparar histograma com a pdf teórica
"""

import numpy as np
import matplotlib.pyplot as plt

# ==============================
# a) Verificação da pdf
# ==============================
# Integral de 2x de 0 a 1 = 1 → válida

print("a) A integral de f(x) = 2x no intervalo [0,1] é 1 → pdf válida")

# ==============================
# b) Função distribuição acumulada
# ==============================
# F(x) = ∫ 2t dt = x^2

print("\nb) F(x) = x^2")

# ==============================
# c) Método da inversa
# ==============================
# Se U ~ Uniform(0,1)
# X = F^{-1}(U) = sqrt(U)

def gerar_amostras(N):
    u = np.random.uniform(0, 1, N)
    x = np.sqrt(u)
    return x

# ==============================
# d) Simulação e comparação
# ==============================
N = 1000
amostras = gerar_amostras(N)

# Valores teóricos
x_teorico = np.linspace(0, 1, 100)
f_teorico = 2 * x_teorico

# Gráfico
plt.hist(amostras, bins=30, density=True)
plt.plot(x_teorico, f_teorico)

plt.title("Histograma vs PDF Teórica")
plt.xlabel("x")
plt.ylabel("Densidade")

plt.show()
```

## Resultados  
---  

### a) Verificação da densidade

\[
\ ∫ [0, 1] 2x dx = 1
\]

Portanto, a função \( f(x) = 2x \) é uma **densidade de probabilidade válida**.

---

### b) Função distribuição acumulada (CDF)

\[
F(x) = x^2
\]

---

### c) Transformação inversa

A partir de \( U ~ Uniform(0,1) \):

\[
X = F^{-1}(U) = sqrt(U)
\]

---

### d) Comparação entre simulação e pdf teórica

Foram geradas 1000 amostras utilizando o método da inversa.

O histograma obtido apresenta boa concordância com a função densidade teórica,
conforme observado no gráfico:

- A forma crescente do histograma acompanha a função \( f(x) = 2x \)
- Pequenas diferenças são devidas à variabilidade estatística da amostra
